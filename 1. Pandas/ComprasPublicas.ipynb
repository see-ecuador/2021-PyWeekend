```
df_publicadas = pd.read_excel("ContratacionesPublicadas.xlsx", skiprows=14,header=1)
```

```
df_publicadas.head(3)
```

```
df_publicadas.columns
```

```
df_publicadas = df_publicadas[['Entidad contratante','Objeto contractual','Monto contratado','Proveedor','Provincia','Cantón']]

df_publicadas.head()
```

```
df_publicadas.describe(include="all")
```

```
df_publicadas.isna().sum()
```

```
df_publicadas[df_publicadas.isnull().any(1)]
```

```
df_publicadas = df_publicadas.drop(903)
```

```
df_publicadas.isna().sum()
```

```
df_provincia =df_publicadas.groupby(['Provincia','Cantón',
                                     'Entidad contratante']).agg(Suma_montos =("Monto contratado",'sum'))
df_provincia.sort_values("Provincia",ascending=True)
```

```
#Vemos los totales por provinvia
df_topProvincia = df_provincia.groupby("Provincia").sum("Suma_montos").sort_values("Suma_montos",ascending=False)

df_topProvincia
```

```
#sacamos el top 5 de provincias
df_topProvincia5 = df_topProvincia.sort_values("Suma_montos",ascending=False).head(5)
df_topProvincia5.sort_values("Suma_montos",ascending=False).head(5)
```

```
#Seaborn barplot
sns.set_style("whitegrid")
bar,ax = plt.subplots(figsize=(10,6))
ax = sns.barplot(x=df_topProvincia5.index, y='Suma_montos', data=df_topProvincia5, ci=None, palette="muted",orient='v', )
ax.set_title("Top 5 de Provincias que más contrataron en Emergencia", fontsize=15)
ax.set_xlabel ("")
ax.set_ylabel ("Monto Contratado")

# Etiqueto 
for p in ax.patches:
 # get the height of each bar
 height = round(p.get_height(),2) 
 # adding text to each bar
 ax.text(x = p.get_x()+(p.get_width()/2), y = height+100,
 s = '{:,}'.format(height), ha = 'center')
```

# **Plotly Express**

```
#Filtramos datos del cantón Guayaquil
df_Guayaquil = df_publicadas.query('Cantón == "GUAYAQUIL"')
df_Guayaquil.columns
```

```
#agrupamos por Entidad contratante y sumamos sumamos los valores contratados
df_Guayaquil5 = df_Guayaquil.groupby("Entidad contratante").sum().sort_values("Monto contratado",ascending=False).head(5)

df_Guayaquil5 = df_Guayaquil5.reset_index()

df_Guayaquil5
```

```
import plotly.express as px
fig = px.pie(df_Guayaquil5, values='Monto contratado', names='Entidad contratante'
, title='Entidades Públicas con mayores montos de contratación en Guayaquil')
fig.show()
```

# **Wordcloud**

https://towardsdatascience.com/simple-wordcloud-in-python-2ae54a9f58e5

```
#!pip install WordCloud

from wordcloud import WordCloud, STOPWORDS
# Create stopword list:

# Generate a word cloud image
text = " ".join(review for review in df_publicadas["Objeto contractual"])

wordcloud = WordCloud(random_state=1,stopwords = STOPWORDS).generate(text)

bar,ax = plt.subplots(figsize=(15,10))
# Display the generated image:
# the matplotlib way:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```
#nltk librería de análisis de lenguaje
import nltk
#Este proceso puede hacerse antes de forma manual, descargar las stopwords de la librería nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words_sp = set(stopwords.words('spanish'))
```

```
wordcloud = WordCloud(random_state=2,stopwords = stop_words_sp,colormap='Set2', collocations=False).generate(text)

bar,ax = plt.subplots(figsize=(15,10))
# Display the generated image:
# the matplotlib way:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
